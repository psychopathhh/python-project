```python
# ======================
# 1. Импорт библиотек
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Настройка стиля графиков
sns.set_theme(style="whitegrid") 
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# ======================
# 2. Загрузка данных
# ======================
try:
    df = pd.read_csv('Hotel Reservations.csv')
    print("Данные успешно загружены!")
    print(f"Размерность данных: {df.shape}\n")
    print("Первые 5 строк данных:")
    display(df.head())
except FileNotFoundError:
    print("Ошибка: Файл 'Hotel Reservations.csv' не найден!")
    exit()

# ======================
# 3. Предварительный анализ
# ======================
print("\nИнформация о данных:")
print(df.info())

print("\nСтатистика числовых данных:")
display(df.describe())

print("\nРаспределение категориальных данных:")
display(df.describe(include=['object']))

# Визуализация распределения целевой переменной
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='booking_status', data=df)
plt.title('Распределение статусов бронирования', pad=20)
plt.xlabel('Статус бронирования')
plt.ylabel('Количество')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.tight_layout()
plt.show()

# ======================
# 4. Предобработка данных
# ======================
# Проверка на пропуски
print("\nПропущенные значения:")
print(df.isnull().sum().to_frame(name='Количество пропусков'))

# Удаление ненужных столбцов
df = df.drop(['Booking_ID'], axis=1)

# Преобразование категориальных переменных
cat_cols = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type', 'booking_status']
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Разделение на признаки и целевую переменную
X = df.drop('booking_status', axis=1)
y = df['booking_status']

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# ======================
# 5. Анализ методом главных компонент (PCA)
# ======================
# Применение PCA для анализа
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Визуализация объясненной дисперсии
plt.figure(figsize=(12, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Количество компонент', fontsize=12)
plt.ylabel('Накопленная объясненная дисперсия', fontsize=12)
plt.title('График объясненной дисперсии PCA', fontsize=14, pad=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(0, len(pca.explained_variance_ratio_)+1, 2))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.tight_layout()
plt.show()

# Определение оптимального числа компонент
n_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
print(f"\nОптимальное количество компонент: {n_components}")

# Повторный PCA с оптимальным числом компонент
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# ======================
# 6. Визуализация результатов
# ======================
# График счетов (первые две компоненты)
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis', s=80)
plt.xlabel('Главная компонента 1', fontsize=12)
plt.ylabel('Главная компонента 2', fontsize=12)
plt.title('График счетов PCA (цвет - статус бронирования)', fontsize=14, pad=20)
plt.legend(title='Статус бронирования', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# График нагрузок (вклад признаков)
plt.figure(figsize=(12, 10))
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
sns.barplot(x=loadings[:, 0], y=X.columns, palette='coolwarm')
plt.xlabel('Вклад в первую главную компоненту', fontsize=12)
plt.ylabel('Признаки', fontsize=12)
plt.title('Вклад признаков в первую главную компоненту', fontsize=14, pad=20)
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ======================
# 7. Дополнительный анализ (кластеризация)
# ======================
# Применение K-means для кластеризации
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Визуализация кластеров
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis', s=80)
plt.xlabel('Главная компонента 1', fontsize=12)
plt.ylabel('Главная компонента 2', fontsize=12)
plt.title('Кластеризация методом K-means', fontsize=14, pad=20)
plt.legend(title='Кластер', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ======================
# 8. Анализ важности признаков
# ======================
# Создание DataFrame с нагрузками
loadings_df = pd.DataFrame(loadings, 
                         columns=[f'PC{i+1}' for i in range(n_components)], 
                         index=X.columns)

print("\nВклад признаков в главные компоненты:")
display(loadings_df.style.background_gradient(cmap='coolwarm', axis=0))

top_features = loadings_df['PC1'].abs().sort_values(ascending=False).head(5)
print("\nТоп-5 наиболее важных признаков:")
display(top_features.to_frame(name='Вклад в PC1'))

# ======================
# 9. Выводы и рекомендации
# ======================
print("\nОсновные выводы:")
print(f"1. Наибольший вклад в отмену бронирований вносят: {top_features.index.tolist()}")
print("2. График счетов показывает четкое разделение данных на кластеры")
print(f"3. Оптимальное число главных компонент: {n_components} (объясняет 95% дисперсии)")

print("\nРекомендации для отеля:")
print("1. Особое внимание следует уделить:", ', '.join(top_features.index[:3]))
print("2. Рассмотреть возможность изменения политики по этим параметрам")
print("3. Использовать PCA-анализ для сегментации клиентов и персонализации предложений")
print("4. Внедрить систему мониторинга ключевых факторов, влияющих на отмены")
```
